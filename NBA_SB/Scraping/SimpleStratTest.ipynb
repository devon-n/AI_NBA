```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Games and Stats.csv')
# data = pd.read_csv('MLB Games.csv')
```

```python
conditions = [
    (data['Home Win'] == False),
    (data['Home Win'] == True)
    ]

values = [
    ((data['Home Odds']-1) * 5),
    (-5)
    ]

data['Outcome'] = np.select(conditions, values)
data['Bankroll'] = data['Outcome'].cumsum()
data['Bankroll'][len(data)-1]
```

```python
plt.plot(data['Bankroll'])
plt.show()
```
