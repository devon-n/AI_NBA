```python
import pandas as pd
import numpy as np
```

```python
years = [i for i in range(2009, 2020)]
```

```python
for year in years:
    stats = pd.read_csv(f'./Stats/NBA {year - 1} Stats.csv')
    games = pd.read_csv(f'./Games/NBA {year} Games.csv')

    # Create home and visitor stats dfs
    home_stats = stats.copy().add_prefix('H ')
    vis_stats = stats.copy().add_prefix('V ')
    home_stats.rename({'H Team': 'Home'}, axis=1, inplace=True)
    vis_stats.rename({'V Team': 'Visitor'}, axis=1, inplace=True)

    # Merge each game with the stats from the prev year
    merged = games.merge(home_stats, on='Home').merge(vis_stats, on='Visitor')

    # concat merged with all_games
    all_games = pd.concat([all_games, merged])

all_games.sort_values(by='Date', inplace=True)
```
