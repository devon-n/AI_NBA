```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

years = [i for i in range(2007, 2021)]



for year in years:
    url = f'https://www.basketball-reference.com/leagues/NBA_{year}.html'

    r = requests.get(url)

    soup = BeautifulSoup(r.content, 'html.parser')

    points_table = soup.find_all('tbody')
    
    stats = []
    
    for table in points_table:
        rows = table.find_all('tr', class_='full_table')
        for row in rows:
            name = row.find('th', class_='left').text.strip('*')
            stat = row.find_all('td', class_='right')
            win = stat[0].text
            loss = stat[1].text
            point = stat[4].text
            point_against = stat[5].text
            
            
            name = name.replace('(', '').replace('*', '')

            # Team name changes
            if name == 'New Orleans Hornets' or name == 'New Orleans/Oklahoma City Hornets':
                name = 'New Orleans Pelicans'
            if name == 'Charlotte Bobcats':
                name = 'Charlotte Hornets'
            if name == 'New Jersey Nets':
                name = 'Brooklyn Nets'
            if name == 'Seattle SuperSonics':
                name = 'Oklahoma City Thunder'
            
            stats_dict = {
                'Team': name.strip(),
                f'{year} Wins': int(win),
                f'{year} Losses': int(loss),
                f'{year} Points For': float(point),
                f'{year} Points Against': float(point_against)
            }
            
            stats.append(stats_dict)

    if year == years[0]:
        stats_df = pd.DataFrame(stats)
    else:
        current_year = pd.DataFrame(stats)
        stats_df = stats_df.merge(current_year, on='Team')
        stats_df.drop_duplicates(subset='Team', keep='first', inplace=True)
            
stats_df
```

```python
stats_df.drop_duplicates(subset='Team', keep='first', inplace=True)
stats_df.reset_index(inplace=True)
stats_df.drop('index', axis=1, inplace=True)
stats_df
```

```python
# Add in the calculations from Machine Learning Sports Betting
merged = stats_df

for year in years:

    # Find total games
    merged[f'{year} Games'] = merged[f'{year} Wins'] + merged[f'{year} Losses']

    # Wins and Losses / Games
    merged[f'{year} W/G'] = merged[f'{year} Wins'] / merged[f'{year} Games']
    merged[f'{year} L/G'] = merged[f'{year} Losses'] / merged[f'{year} Games']

    # Find margin (Points for - Points against)
    merged[f'{year} Margin'] = merged[f'{year} Points For'] - merged[f'{year} Points Against']

    # Rank (W, L, GF, GA, M)
    merged[f'{year} WRank'] = merged[f'{year} Wins'].rank(ascending=False)
    merged[f'{year} LRank'] = merged[f'{year} Losses'].rank(ascending=False)
    merged[f'{year} GFRank'] = merged[f'{year} Points For'].rank(ascending=False)
    merged[f'{year} GARank'] = merged[f'{year} Points Against'].rank(ascending=False)
    merged[f'{year} MRank'] = merged[f'{year} Margin'].rank(ascending=False)

    # Power Ranking
    merged[f'{year} WPower'] = (1+(1-(merged[f'{year} WRank']/merged[f'{year} WRank'].mean())))*0.5
    merged[f'{year} LPower'] = (1+(1-(merged[f'{year} LRank']/merged[f'{year} LRank'].mean())))*0.5
    merged[f'{year} GFPower'] = (1+(1-(merged[f'{year} GFRank']/merged[f'{year} GFRank'].mean())))*0.5
    merged[f'{year} GAPower'] = (1+(1-(merged[f'{year} GARank']/merged[f'{year} GARank'].mean())))*0.5
    merged[f'{year} MPower'] = (1+(1-(merged[f'{year} MRank']/merged[f'{year} MRank'].mean())))*0.5
```
